# Continents



```python
# get table headers
import pandas as pd

continents = pd.DataFrame(columns = ['Country', 'Capital', 'Population', 'Area', 'Continent'])
names = ['Africa', 'Asia', 'Europe', 'North_America', 'South_America', 'Oceania']

for i in range(6):
  df = pd.read_html('https://simple.wikipedia.org/wiki/List_of_countries_by_continents', header=0)[i]
  df.drop(df.columns[0], axis=1, inplace=True)
  df.drop(df.columns[-1], axis=1, inplace=True)
  if(df.columns[0]=='Flag'):
    df.drop(['Flag'], axis=1, inplace=True)

  df = df.iloc[:,:4]
  df.columns = ['Country', 'Capital', 'Population', 'Area']
  df['Continent'] = names[i]
  df = df.astype(str)
  continents = pd.concat([continents, df])

print(continents.shape)
print(continents.info())
print(continents['Continent'].value_counts())
continents.to_csv('continents.csv')
continents.head()
```
