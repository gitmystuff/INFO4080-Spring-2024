# Melt



```python
import pandas as pd

df = pd.read_csv('https://docs.google.com/spreadsheets/d/18tZbDl8q8u3hktyWHQYDjprn8KUKqK4AgZTKXU4J0gM/export?format=csv')
df.head()
```

```python
df.describe()
```

```python
world_pop = df.melt(id_vars=['Country Name'], value_vars=['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'])
world_pop.columns = ['Country', 'Year', 'Population']
world_pop.to_csv('wrld_pop.csv', index=False)
world_pop.head()
```
